# Tic Tac Toe Game

```python
import numpy as np
```

```python
class Game:
    player = ('O', 'X')
    N = 3
    
    def __init__(self, p1, p2, computer_first=False):
        self.computer_turn = computer_first
        self.human = p1
        self.bot = p2
        self.board = [str(i+1) for i in range(self.N**2)]
        
    def start(self):
        rnd = 1

        while rnd <= self.N**2:
            player = self.player[self.computer_turn]
            print("= Round %d =" % rnd)
            
            if self.computer_turn:
                pos = self.bot.move(self.board) 
            else:
                pos = self.human.move(self.board)
            
            self.board[pos] = player
            self.print_board(self.board)
                
            if self.is_win(self.board, player):
                if self.computer_turn:
                    print("You lose!!!")
                else:
                    print("You win!!!")
                break

            self.computer_turn = not self.computer_turn
            rnd += 1
    
    def print_board(self, board):
        cur_board = list(board)
        for i, c in enumerate(cur_board):
            if c == 'O':
                cur_board[i] = '\033[1;31mO\033[0m'
            if c == 'X':
                cur_board[i] = '\033[1;34mX\033[0m'
        
        print()
        print('  %s  |  %s  |  %s' % (cur_board[0], cur_board[1], cur_board[2]))
        print(' ----|-----|----')
        print('  %s  |  %s  |  %s' % (cur_board[3], cur_board[4], cur_board[5]))
        print(' ----|-----|----')
        print('  %s  |  %s  |  %s' % (cur_board[6], cur_board[7], cur_board[8]))
        print()
        
        
    def is_win(self, board, player):
        win_state = (
            (0,1,2), (3,4,5), (6,7,8),
            (0,3,6), (1,4,7), (2,5,8),
            (0,4,8), (2,4,6)
        )
        for s in win_state:
            i, j, k = s
            if player == board[i] == board[j] == board[k]:
                return True
        return False
```

```python
class Player:
    def __init__(self, name="human"):
        self.name = name
        
    def move(self, board):
        print(">> You turn")
        pos = int(input("Please Enter number: ")) - 1
        if not self.valid_move(board, pos):
            pos = int(input("Please Enter number: ")) - 1
        return pos
    
    def valid_move(self, board, pos):
        return not board[pos] in 'OX'
```

```python
class Bot(Player):
    def __init__(self, name="bot"):
        self.name = name
        
    def move(self, board):
        return self.minimax(board)
    
    def minimax(self, board):
        max_list = self.max_val(list(board))

        if max_list[0][0] == 0:
            max_list = [k[1] for k in max_list if k[0] == 0]
            return max_list[np.random.randint(0, len(max_list))]

        return max_list[0][1]
    
    def min_val(self, board):
        min_list = list()
        for j, is_space_o in enumerate(board):
            if is_space_o not in 'OX':
                temp_board = list(board)
                temp_board[j] = 'O'

                score = 0
                if self.is_win(temp_board, 'X'):
                    score = 10
                elif self.is_win(temp_board, 'O'):
                    score = -10
                min_list.append((score, j))

        min_list.sort()
        return min_list
    
    def max_val(self, board):
        max_list = list()
        for i, is_space_x in enumerate(board):
            if is_space_x not in 'OX':
                cur_board = list(board)
                cur_board[i] = 'X'
                min_list = self.min_val(board)
                   
                max_list.append(min_list[0])
        max_list.sort(reverse=True)
        return max_list
    
    def is_win(self, board, player):
        win_state = (
            (0,1,2), (3,4,5), (6,7,8),
            (0,3,6), (1,4,7), (2,5,8),
            (0,4,8), (2,4,6)
        )
        for s in win_state:
            i, j, k = s
            if player == board[i] == board[j] == board[k]:
                return True
        return False
```

```python
p1 = Player()
p2 = Bot()

game = Game(p1, p2)
game.start()
```
